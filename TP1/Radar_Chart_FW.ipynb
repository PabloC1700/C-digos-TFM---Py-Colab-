```python
import pandas as pd
import matplotlib.pyplot as plt
!pip install soccerplots
from soccerplots.radar_chart import Radar
```

```python
#Read in the data
df = pd.read_csv('/content/Big_5_European_Leagues_Stats (1).csv')
```

```python
df.head()
```

```python
# Replace the headers with the first row
df.columns = df.iloc[0]
df = df[1:]

# Reset the index
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split('  ',expand=True)[0]
```

```python
df.head()
```

```python
df.Player.unique()
```

```python
df = df[(df['Player']=='Deniz Undav') | (df['Player']=='Artem Dovbyk') | (df['Player']=='Jonathan David') | (df['Player']=='Omar Marmoush')].reset_index()
```

```python
df
```

```python
df = df.drop(['index','Rk','Nation','Pos','Squad','Comp','Age','Born','90s','FK','PK','PKatt','Matches'],axis=1)
```

```python
df
```

```python
#get parameters
params = list(df.columns)
params = params[1:]
params
```

```python
#add ranges to list of tuple pairs
ranges = []
a_values = []
b_values = []
c_values = []
d_values = []

for param in params:
    a = df[param].astype(float).min()
    a = a - (a * 0.25)
    b = df[param].astype(float).max()
    b = b + (b * 0.25)

    ranges.append((a,b))

for x in range(len(df['Player'])):
    if df['Player'][x] == 'Deniz Undav':
      a_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Artem Dovbyk':
      b_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Jonathan David':
      c_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Omar Marmoush':
      d_values = df.iloc[x].values.tolist()


a_values = a_values[1:]
b_values = b_values[1:]
c_values = c_values[1:]
d_values = d_values[1:]

values = [a_values, b_values, c_values, d_values]
```

```python
values
```

```python
print("Ranges:", ranges)
print("Params:", params)
print("a_values:", a_values)
print("b_values:", b_values)
print("c_values:", c_values)
print("d_values:", d_values)

print("Types in ranges:", [(type(a), type(b)) for a, b in ranges])
print("Types in params:", [type(p) for p in params])
print("Types in a_values:", [type(v) for v in a_values])
print("Types in b_values:", [type(v) for v in b_values])
print("Types in c_values:", [type(v) for v in c_values])
print("Types in d_values:", [type(v) for v in d_values])
```

```python
ranges = [(float(a), float(b)) for a, b in ranges]
```

```python
a_values = [float(i) for i in a_values if pd.notnull(i)]
b_values = [float(i) for i in b_values if pd.notnull(i)]
c_values = [float(i) for i in c_values if pd.notnull(i)]
d_values = [float(i) for i in d_values if pd.notnull(i)]
```

```python
#title

title = dict(
    title_name='Comparación de Delanteros',
    title_color = 'red',
    subtitle_name = '5 Grandes Ligas para AC Milan',
    subtitle_color = 'red',
    title_fontsize = 18,
    subtitle_fontsize = 15,
)

endnote = 'Pablo Crespo \nvia Fbref'
```

```python
# Convertir todos los valores a numérico si aún no lo son
values = [list(map(float, v)) for v in values]

# Asegurar que los valores de rango sean numéricos
range_values = [(0, 100)] * len(params)
```

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges,params=params,values=values,
                          radar_color=['black','red','blue','yellow'],
                          alphas=[.75,.6,.5,.4],title=title,endnote=endnote,
                          compare=True)
```
